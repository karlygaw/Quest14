## ЗАДАНИЕ 1: Создаём данные на основе уравнения

### Задание 1, часть 1: Подготовка данных

Важно уметь отобразить данные уравнения на графике. Первая задача будет на самом деле относиться к numpy, и лишь затем мы будет отрисовывать эти данные с помощью Matplotlib. К качестве уравнения мы будем использовать [очень известное уравнение](https://ru.wikipedia.org/wiki/%D0%AD%D0%BA%D0%B2%D0%B8%D0%B2%D0%B0%D0%BB%D0%B5%D0%BD%D1%82%D0%BD%D0%BE%D1%81%D1%82%D1%8C_%D0%BC%D0%B0%D1%81%D1%81%D1%8B_%D0%B8_%D1%8D%D0%BD%D0%B5%D1%80%D0%B3%D0%B8%D0%B8):

$$E=mc^2$$


С помощью Ваших знаний Numpy создайте два массива: E и m:
- значения **m** - это 11 равноотстоящих друг от друга чисел от 0 граммов до 10 граммов
- значения **E** должны быть равны энергии для этих масс.   

Вам нужно будет указать значение для **c**, приведя его в нужные единицы измерения - в метрах в секунду. Можете найти ответ с помощью поиска в Google (достаточно будет указать примерное значение).

**ЗАМЕЧАНИЕ: Если Вам непонятна постановка задачи, то не стесняйтесь посмотреть решение в блокноте с решениями.**

```python
import numpy as np

m = np.linspace(0, 10, 11)
print(m)
```

```python
c = 3e8  # скорость света в метрах в секунду
E = m * c**2
```

```python
print(E)
```

### Задание 1, часть 2: Нарисовать график E=mc^2

Теперь, когда у нас есть массивы E и m, мы можем нарисовать график зависимости энергии от массы. 

**Задание: выполните команду Import для рисования графиков в Matplotlib:**

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

**Задание: воспроизведите показанный ниже график, используя ранее созданные массивы. Обратите внимание на цвета, оси и надписи. Необязательно создавать точно такой же график, но постарайтесь воспроизвести все основные элементы графика.**

```python
# ПИШИТЕ ВАШ КОД ЗДЕСЬ
plt.figure(figsize=(5, 4))  # изменение размера графика
plt.plot(m, E, color='red', linewidth=5)
plt.title('E=mc^2')
plt.xlabel('Mass in Grams')
plt.ylabel('Energy in Joules')
plt.show()
```

**Можете ли Вы выяснить, как при рисовании этого графика получить на оси y логарифмическую шкалу? Также нанесите сетку (grid) по оси y. "Кривая доходности" в финансах - это график, показывающий доходность или процентные ставки в зависимости от длины контракта (2 месяца, 2 года, 20 лет и т.д.). Проценты в долларах, выплачиваемые по ценным бумагам казначейства США (U.S. Treasury securities) для разных сроков погашения, представляют интерес для трейдеров, и такой график неформально называют "the yield curve" - кривая доходности.

**В этом упражнении у Вас будут данные для кривой доходности для двух различных моментов времени. И Вы построите графики на основе этих данных.**

## Задание 2, часть 1: Данные для кривой доходности

**Мы взяли для Вас нужные данные для кривой доходности с сайта [US Treasury Dept.](https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/textview.aspx?data=yield) Данные показывают процентные ставки облигаций США для различных сроков контракта. Список labels отображает сроки контракта для соответствующих позиций в списках значений.**

**Задание: Запустите ячейку ниже, чтобы создать списки для последующего рисования графика.**

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

**Задание: выясните, как можно нарисовать обе эти линии на одном и том же графике. Добавьте на график легенду.**

```python
# ПИШИТЕ ВАШ КОД ЗДЕСЬ
# Строим графики для обоих наборов данных
plt.figure(figsize=(5, 4))
plt.plot(labels, july16_2007, label='July 16, 2007')
plt.plot(labels, july16_2020, label='July 16, 2020')

# Добавляем легенду
plt.legend(fontsize='small') #шрифт меньше сделала чтобы был похож на графика ниже

# Показываем график
plt.show()
```

```python
# НЕ ЗАПУСКАЙТЕ ЯЧЕЙКУ НИЖЕ, ИНАЧЕ ГРАФИК ИСЧЕЗНЕТ!
```

**Задание: на графике выше легенда выглядит несколько странно - хотя она и не загораживает график, было бы лучше поместить легенду *снаружи* графика. Найдите способ переместить легенду так, чтобы она оказалась снаружи.**

```python
# ПИШИТЕ ВАШ КОД ЗДЕСЬ
plt.figure(figsize=(6, 4))
plt.plot(labels, july16_2007, label='July 16, 2007')
plt.plot(labels, july16_2020, label='July 16, 2020')

# Добавляем легенду
plt.legend(fontsize='small', bbox_to_anchor=(1.35, 0.5), loc='center right')


# Показываем график
plt.show()
```

```python
# НЕ ЗАПУСКАЙТЕ ЯЧЕЙКУ НИЖЕ, ИНАЧЕ ГРАФИК ИСЧЕЗНЕТ!
```

**Задание: хотя на графике ясно видно разницу между 2007 и 2020 годами, когда мы наносим эти линии на один и тот же график, нам становится сложнее увидеть детали внутри каждого отдельного года. С помощью .suplots() нарисуйте графики ниже, в которых эти детали будут лучше видны.**

```python
# ПИШИТЕ ВАШ КОД ЗДЕСЬ
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(7, 5), sharex=True)

# Первый график для 2007 года
line1, = axes[0].plot(labels, july16_2007, label='July 16, 2007')
axes[0].set_title('July 16, 2007')

# Второй график для 2020 года
line2, = axes[1].plot(labels, july16_2020, label='July 16, 2020')
axes[1].set_title('July 16, 2020')

# Добавляем легенду
axes[0].legend(handles=[line1], loc='upper left')
axes[1].legend(handles=[line2], loc='upper left')

# Показываем графики
plt.tight_layout()
plt.show()
```

```python
# НЕ ЗАПУСКАЙТЕ ЯЧЕЙКУ НИЖЕ, ИНАЧЕ ГРАФИК ИСЧЕЗНЕТ!
```

**Дополнительное бонусное задание: попробуйте воспроизвести график ниже, с двумя осями. Хотя в данном примере две разные оси по сути не нужны, это хорошее упражнение по Matplotlib, которое пригодится в других ситуациях.**

```python
# ПИШИТЕ ВАШ КОД ЗДЕСЬ
# Создаем фигуру и оси
fig, ax1 = plt.subplots(figsize=(8, 6))

# Первый график
line1, = ax1.plot(labels, july16_2007, color='blue', label='July 16, 2007')
ax1.set_xlabel('Maturity')
ax1.set_ylabel('2007', color='blue')
ax1.tick_params('y', colors='blue')

# Создаем вторые оси
ax2 = ax1.twinx()
line2, = ax2.plot(labels, july16_2020, color='red', label='July 16, 2020')
ax2.set_ylabel('2020', color='red')
ax2.tick_params('y', colors='red')

# Добавляем легенду
ax1.legend(handles=[line1], loc='upper left')
ax2.legend(handles=[line2], loc='upper right')

# Показываем график
plt.show()

```
